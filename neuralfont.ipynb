## postprocessing

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
import os
path = '/content/drive/MyDrive/neural-fonts-master/newfonts/save'
files = os.listdir(path)

denoising_path = '/content/drive/MyDrive/neural-fonts-master/newfonts/denoising_3'
os.mkdir(denoising_path)

reduction_path = '/content/drive/MyDrive/neural-fonts-master/newfonts/reduction_3'
os.mkdir(reduction_path)


both_path = '/content/drive/MyDrive/neural-fonts-master/newfonts/both_3'
os.mkdir(both_path)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
for file in files:
  img = cv2.imread(f'{path}/{file}')
  de_img = cv2.fastNlMeansDenoisingColored(img,None,50,10,7,21)
  re_img = cv2.dilate(img, kernel)

  result_img = cv2.fastNlMeansDenoisingColored(re_img,None,50,10,7,21)

  cv2.imwrite(f'{denoising_path}/{file}',de_img)
  cv2.imwrite(f'{reduction_path}/{file}',re_img)
  cv2.imwrite(f'{both_path}/{file}',result_img)
```
